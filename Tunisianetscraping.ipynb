```python
#Automation
def import_all_data_from_one_page(site):
  responses = requests.get(site,verify=False)
  soups = BeautifulSoup(responses.content, 'html.parser')
  results = soups.find_all('h2', {'class':'h3 product-title'}) # ==> we have all info from one page
  liste = []
  for result in results:
    one_pc_link = result.find('a', href=True).get("href")
    response = requests.get(one_pc_link,verify=False)
    soup = BeautifulSoup(response.content, 'html.parser')
    result_pc = soup.find_all('section', {'class':'product-features'})
    #get pc technical info
    info_pc = {}
    for dt,dd in zip(result_pc[0].find_all('dt'),result_pc[0].find_all('dd')):
       info_pc[dt.text]=dd.text
    
    #get pc price
    price = soup.find_all('span', {'itemprop':'price'})
    info_pc["price"] = price[0].text

    #get pc name
    info_pc['name'] = result.text.split('/')[0]
     
    #add pc link
    info_pc['link'] = one_pc_link
    
    #get pc model
    info_pc['model'] = one_pc_link.split("-")[5]
  
    liste.append(info_pc)

  return liste
  
```

```python
data_one_page = import_all_data_from_one_page("http://www.tunisianet.com.tn/702-ordinateur-portable")
```

```python
import pandas as pd
data = pd.DataFrame(data_one_page)
data
```

```python
#Extract all data : scraping multiples pages
url = "https://www.tunisianet.com.tn/702-ordinateur-portable?page="
for i in range(2,38):
  data = pd.concat([data, pd.DataFrame(import_all_data_from_one_page(url+str(i)))])
```
